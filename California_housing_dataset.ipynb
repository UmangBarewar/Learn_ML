```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

 ### Choosing the right Estimator/algorithm for your problem 

###### Let's use California housing datset from =https://scikit-learn.org/stable/modules/generated/sklearn.datasets.fetch_california_housing.html#sklearn.datasets.fetch_california_housing

```python
## Getting data set for regression
from sklearn.datasets import fetch_california_housing
housing = fetch_california_housing()
housing
```

From here we can clearly see observe that, the data is stored in dictionary format, which has key "data", "target","target_names" & "feature_names" in there. Feauture_names has total 7 columns (axis=1), and target_names has only one 1 columns which is "MedHouseVal". 

```python
housing_df=pd.DataFrame(housing["data"],columns=housing["feature_names"])
housing_df
```

Here we are creating dataframes (housing_df) form the key "data" from the dataset which has been stored in the form of Dictionary. These values under "data" key will be stored in columns whose name is taken from another key "feature_names" which has total 7 features.

```python
housing_df["MedHouseVal"]=housing["target"]
housing_df.head()
```

```python
housing_df["target"]=housing["target"]
housing_df.head()
```

```python
housing_df=housing_df.drop("MedHouseVal",axis=1)
housing_df
```

```python
#Importing the algorithm/estimatore
from sklearn.linear_model import Ridge
from sklearn.model_selection import train_test_split
#Setting up random sees
np.random.seed(42)

#Creating the data
X=housing_df.drop("target",axis=1)
y=housing_df["target"]
#Split into train and test sets
X_train, X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
#Instantiating and fitting the model on training set
model=Ridge()
model.fit(X_train,y_train)
#Checking the model score on the testing set
model.score(X_test,y_test)
```

```python
# Import the RandomForestRegressor model class
from sklearn.ensemble import RandomForestRegressor
Rf_model=RandomForestRegressor()
Rf_model.fit(X_train,y_train)

Rf_model.score(X_test,y_test)
```

```python
y_preds=Rf_model.predict(X_test)
```

### Mean Absolute Error

```python
df=pd.DataFrame(data={"Actual Values":y_test,
                     "Predicted Values":y_preds})
df[" Absolute Differences"]=abs(df["Actual Values"]-df["Predicted Values"])
df["Differences"]=(df["Actual Values"]-df["Predicted Values"])
df.head(20)
```

```python
df["Differences"].mean()
```

```python
# Compare the prediction to the truth
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_test,y_preds)
```

Mean absolute error simply means the the mean of difference between the y_preds and y_test values

### Mean Squared Error
Mean Squared Error : Mean of the square of the errors between actual
predicted values

```python
df["Squared Differences"]=np.square(df["Differences"])
df.head()
```

```python
MSE=df["Squared Differences"].mean()
MSE
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test,y_preds)
```

### R2 score

```python
from sklearn.metrics import r2_score
# Fill an array with y_test mean
y_test_mean=np.full(len(y_test),y_test.mean())
```

```python
y_test_mean[:10]
```

```python
r2_score(y_test,y_test)
```

What R ^2 does iscompare your model prediction to the mean of targets. Values can range from negative infinity to one. For example, if all of your model does predict the mean of targets, its R Square value would be zero. And if your model perfect predicts a range of number, its R Square value would be 1.
