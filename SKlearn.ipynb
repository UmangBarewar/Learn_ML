```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

## An end to end scikit learn workflow

```python
heart_disease=pd.read_csv("heart-disease.csv")
heart_disease
```

```python
# import warnings
# warnings.filterwarnings("ignore")
#warnings.filterwarnings("default")
```

```python
#1. create X (features matrix)
X=heart_disease.drop("target",axis=1)
#create Y (labels)
y=heart_disease["target"]
```

```python
#2.choose the right model and hyperparameters
from sklearn.ensemble import RandomForestClassifier
clf =  RandomForestClassifier()
# we will keep the default hyperparameters
clf.get_params();
```

```python
#3. Fit the modelto the training data 
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
clf.fit(X_train,y_train)
```

```python
X_train[:5],X_test[:5]
```

```python
# make a prediction
y_label = clf.predict(np.array([0,1,2,3,4]))
```

```python
y_preds= clf.predict(X_test)
y_preds
```

```python
#4. Evaluate the model
clf.score(X_train,y_train)
```

```python
clf.score(X_test,y_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
print(classification_report(y_test,y_preds))
```

```python
#5. Improve the metal
# try differnt amount of n_estimators
np.random.seed(42)
for i in range(10,100,10):
    print(f"Trying model with {i} estimators")
    clf=RandomForestClassifier(n_estimators =i).fit(X_train,y_train)
    print(f"Model accuracy on test set : {clf.score(X_test,y_test)*100}")
```

```python
#6. Save a model and load it
import pickle

pickle.dump(clf,open("random_forestmodel1.pkl","wb"))
```

```python
load_model=pickle.load(open("random_forestmodel1.pkl","rb"))
load_model.score(X_test,y_test)
```
